```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('surveyofbodyfat.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
#for outliers 
```

```python
missvalues = df.isnull().sum()
missvalues
# so there is no problem about missing values, that means we did not need to add addtion values. go on current dataset
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8)) # default 12 8 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') # colours it is up to you that one is famous
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix[['BodyFat']].sort_values(by='BodyFat', ascending=False), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#that is  the HEATMAP (another way to display the correlation)
```

```python
sns.scatterplot(data=df, x='Density', y='BodyFat')
plt.show()


strong_affects= ['Abdomen', 'Hip', 'Weight', 'Age']


for i in strong_affects:
    plt.figure(figsize=(8, 6)) # 10
    sns.scatterplot(data=df, x=i, y='BodyFat')
    plt.title(f'Scatter Plot of {i} vs BodyFat')
    plt.xlabel(i)
    plt.ylabel('BodyFat')
    plt.show()


sns.histplot(df['BodyFat'], bins=20, kde=True) # kde ? # bins done 
plt.show()
```

```python
moderate_affects = ['Chest', 'Thigh', 'Neck']

for i in moderate_affects:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=i, y='BodyFat')
    plt.title(f'Scatter plot of {i} vs Bodyfat')
    plt.xlabel(i)
    plt.ylabel('BodyFat')
    plt.show()
```

```python

regressions = ['Abdomen', 'Hip', 'Weight', 'Age', 'Chest', 'Thigh', 'Neck']

for i in regressions:
    plt.figure(figsize=(8, 6))
    sns.regplot(data=df, x=i, y='BodyFat', scatter_kws={'s':10}, line_kws={'color':'red'})
    plt.title(f'regression for {i} vs bodyfat')
    plt.xlabel(i)
    plt.ylabel('BodyFat')
    plt.show()
```

```python
# for more detail we can use pairplot it is the simple and popular script in library (sns) you can check What is pairplot in goglee
# so we can describe the relation of each feature with bodyfat in own frame
# it resembles correlation analysis but not the numbers directly graphs

features_subset = ['BodyFat', 'Abdomen', 'Hip', 'Weight', 'Age', 'Chest', 'Thigh', 'Neck']


sns.pairplot(df[features_subset], diag_kind='kde', kind='scatter')
plt.suptitle('pairplot of Features with BodyFat', y=1.02)
plt.show()
```
