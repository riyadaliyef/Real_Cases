```python
import pandas as pd
```

```python
df = pd.read_csv('creditlimit.csv')
```

```python
df_encoded = pd.get_dummies(df, columns=['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category'], drop_first=True)
```

```python
df_encoded.head(10)
```

```python
# it is obvious that i create new 3 df in order to find fraud or not
df_encoded['istifade_faizle'] = df_encoded['Total_Revolving_Bal'] / df_encoded['Credit_Limit']
df_encoded['xercleme'] = df_encoded['Total_Trans_Amt'] / df_encoded['Months_on_book']
df_encoded['kredit_faizle'] = df_encoded['Avg_Open_To_Buy'] / df_encoded['Credit_Limit']
```

```python
df_encoded[['istifade_faizle', 'xercleme', 'kredit_faizle']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# here we can see vizualization of customers and the line shows us averages
plt.figure(figsize=(10, 6))
sns.histplot(df_encoded['istifade_faizle'], bins=50, kde=True)
plt.axvline(df_encoded['istifade_faizle'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.title('Istifadənin faizlə bölgüsü')
plt.xlabel('Faiz')
plt.ylabel('Tezlik')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_encoded['xercleme'], bins=50, kde=True)
plt.axvline(df_encoded['xercleme'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.title('Xerclemenin bölgüsü')
plt.xlabel('xercleme')
plt.ylabel('Tezlik')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_encoded['kredit_faizle'], bins=50, kde=True)
plt.axvline(df_encoded['kredit_faizle'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.title('Kreditin faizlə bölgüsü')
plt.xlabel('Kredit faizle')
plt.ylabel('Tezlik')
plt.show()
```

```python
from scipy.stats import zscore
df_encoded['z_istifade_faizle'] = zscore(df_encoded['istifade_faizle'])
df_encoded['z_xercleme'] = zscore(df_encoded['xercleme'])
df_encoded['z_kredit_faizle'] = zscore(df_encoded['kredit_faizle'])

outliers = df_encoded[(df_encoded['z_istifade_faizle'].abs() > 3) | 
                      (df_encoded['z_xercleme'].abs() > 3) | 
                      (df_encoded['z_kredit_faizle'].abs() > 3)]

outliers.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

df_encoded['fraud'] = ((df_encoded['z_istifade_faizle'].abs() > 3) | 
                          (df_encoded['z_xercleme'].abs() > 3) | 
                          (df_encoded['z_kredit_faizle'].abs() > 3)).astype(int)


X = df_encoded[['istifade_faizle', 'z_xercleme', 'kredit_faizle']]
y = df_encoded['fraud']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```
