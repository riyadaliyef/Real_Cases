```python
# our target
# 1)  the number of each regional hospital 
# 2)  over the age analysis
# 3)   each department's sum of deposit
```

```python
import pandas as pd
```

```python
df = pd.read_csv('hospitaldataset.csv')
df.head()
```

```python
#1 number of patients in the region
```

```python
Sum_of_hospitals = df['Hospital_region_code'].value_counts().reset_index()
Sum_of_hospitals
```

```python
Sum_of_hospitals.columns = ['Hospital_region_code', 'Total']
Sum_of_hospitals
```

```python
#2 age analysis of hospital
```

```python
age_analysis = df['Age'].value_counts().reset_index()
age_analysis
```

```python
age_analysis.columns = ['Age_Group', 'Total']
age_analysis
```

```python
#3 total deposit for each department
```

```python
department_deposits_sum = df.groupby('Department')['Admission_Deposit'].sum().reset_index()
department_deposits_sum
```

```python
department_deposits_sum.columns = ['Department', 'Total(Deposit)']
department_deposits_sum
```
